```python
# Step 1: Import necessary libraries
from ultralytics import YOLO
import cv2  # For image processing
import matplotlib.pyplot as plt  # For displaying images

# Step 2: Define paths
image_path = r"C:\Users\arpan\ISRO_Hackathon\Batch_8_ISRO_Dataset\tile_71_7.png"  # Change this to the path of your image

# Step 3: Load the trained model
model = YOLO('runs/detect/custom_yolov8_model4/weights/best.pt')  # Update the path if necessary

# Step 4: Predict on the image
results = model.predict(source=image_path)

# Step 5: Visualize the predictions
# Convert the image to RGB (cv2 loads images in BGR format)
image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

# Extract the first result from the results list
result = results[0]

# Draw the bounding boxes and labels on the image
for box in result.boxes.xyxy:
    x1, y1, x2, y2 = map(int, box)  # Convert to integer values
    cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Red box with thickness 2

# Extract labels and scores
labels = result.boxes.cls.cpu().numpy()
scores = result.boxes.conf.cpu().numpy()

for box, score, label in zip(result.boxes.xyxy, scores, labels):
    x1, y1, x2, y2 = map(int, box)  # Convert to integer values
    label_text = f"{label}: {score:.2f}"
    cv2.putText(image, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# Display the image with matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Step 1: Import necessary libraries
from ultralytics import YOLO
import cv2  # For image processing
import matplotlib.pyplot as plt  # For displaying images

# Step 2: Define paths
image_path = r"C:\Users\arpan\ISRO_Hackathon\Batch_8_ISRO_Dataset\tile_71_9.png"  # Change this to the path of your image

# Step 3: Load the trained model
model = YOLO('runs/detect/custom_yolov8_model4/weights/best.pt')  # Update the path if necessary

# Step 4: Predict on the image
results = model.predict(source=image_path)

# Step 5: Visualize the predictions
# Convert the image to RGB (cv2 loads images in BGR format)
image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

# Define class names and colors
class_names = {0: 'crater', 1: 'boulder'}
colors = {0: (255, 0, 0), 1: (0, 255, 0)}  # Red for boulder, Green for crater

# Extract the first result from the results list
result = results[0]

# Draw the bounding boxes and labels on the image
for box, label in zip(result.boxes.xyxy, result.boxes.cls.cpu().numpy()):
    x1, y1, x2, y2 = map(int, box)  # Convert to integer values
    color = colors[label]  # Get color based on label
    cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)  # Draw rectangle with the specified color

# Extract labels and scores
labels = result.boxes.cls.cpu().numpy()
scores = result.boxes.conf.cpu().numpy()

for box, score, label in zip(result.boxes.xyxy, scores, labels):
    x1, y1, x2, y2 = map(int, box)  # Convert to integer values
    label_text = f"{class_names[label]}: {score:.2f}"
    color = colors[label]  # Get color based on label
    cv2.putText(image, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# Display the image with matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')
plt.show()
```
